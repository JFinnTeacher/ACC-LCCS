```python
#
import random

number = random.randint(1, 10)
# print(number)

guess = int(input("Enter a Number between 1 and 10: "))

#
if guess == number:
  print("Your guess was correct")
  print("Well Done")
elif guess > number:
  print("That is incorrect")
  print("Your guess was too high")
else:
  print("That is incorrect")
  print("Your guess was too low")

print("Goodbye")
```
