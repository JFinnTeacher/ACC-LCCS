```python
import matplotlib.pyplot as plt

StudentNames = ["John","James","Michelle","Rosie","Paul","Leonard"]
StudentHeight = [1.5,1.4,1.6,1.8,1.4,1.5]
plt.scatter(StudentNames,StudentHeight,marker="s",color="red")
plt.xlabel("Names")
plt.ylabel("Height in M")
plt.title("Student Heights")
plt.show()
```

**TASK 3**

Repeat Question 2 but change the plot type to a bar graph

```python
import matplotlib.pyplot as plt

StudentNames = ["John","James","Michelle","Rosie","Paul","Leonard"]
StudentHeight = [1.5,1.4,1.6,1.8,1.4,1.5]
plt.bar(StudentNames,StudentHeight,color="red")
plt.xlabel("Names")
plt.ylabel("Height in M")
plt.title("Student Heights")
plt.show()
```

```python
import matplotlib.pyplot as plt

name = ["Lionel Messi","Steph Curry","Conor McGregor","Serena Williams","Danica Patrick","Katie Taylor"]
age = [32,31,30,37,37,33]
earnings = [127,79,47,18,7.5,1.1]
plt.plot(name,age, label = "age")
plt.plot(name,earnings, label = "Earnings")
plt.xlabel("Names")
plt.legend()
plt.show()
```
