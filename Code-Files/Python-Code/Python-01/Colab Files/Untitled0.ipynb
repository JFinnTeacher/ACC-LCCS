```python
# Import the tkinter module
import tkinter as tk

# Create a new tkinter window
window = tk.Tk()

# Create a new canvas inside the window
canvas = tk.Canvas(window, width=400, height=400)
canvas.pack()

# Draw a circle on the canvas
# The parameters are: canvas.create_oval(x0, y0, x1, y1)
# (x0, y0) is the top-left coordinate where the circle starts.
# (x1, y1) is the bottom-right coordinate where the circle ends.
# You can also add fill="color_name" to fill the circle with a color.
canvas.create_oval(50, 50, 150, 150)

# Create a dot at (250, 250), which is the center
dot = canvas.create_oval(250, 250, 260, 260, fill='red')

radius = 100  # The radius of orbit
angle = 0     # The initial angle

def rotate_dot():
    global angle

    # Calculate the new position of the dot
    x = 250 + radius * math.cos(angle)
    y = 250 + radius * math.sin(angle)

    # Move the dot to the new position
    canvas.coords(dot, x-5, y-5, x+5, y+5)

    # Increase the angle
    angle += 0.02

    # Repeat every 10 milliseconds
    window.after(10, rotate_dot)

# Start the rotation
window.after(10, rotate_dot)

# Start the tkinter event loop (this will display the window)
window.mainloop()
```
