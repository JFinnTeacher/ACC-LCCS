# Working with Lists

Initilising Lists

```python
# Lists can be initially empty
boysNames = []
girlsNames = []
favouriteSongs = []
fruits = []
vehicleCount = []
accountDetails = []
```

Populating List

```python
# Lists can be created with data (each value is a list element)
boysNames = ['John', 'Jim', 'Alex', 'Fred']
girlsNames = ['Sarah', 'Alex', 'Pat', 'Mary']
favouriteSongs = ['Moondance', 'Linger', 'Stairway to Heaven']
fruits = ['Strawberry', 'Lemon', 'Orange', 'Raspberry', 'Cherry']
vehicleCount = [0, 0, 0, 0, 0, 0]
accountDetails = [1234, 'xyz', 'Alex', '1 Main Street', 827.56]
```

Joining Lists

```python
# Concatenate two lists and store the result in a new list
boysNames = ['John', 'Jim', 'Alex', 'Fred']
girlsNames = ['Sarah', 'Alex', 'Pat', 'Mary']
names = boysNames + girlsNames
print(names)
```

Concatenate 2 lists by extending one

```python
# Concatenate two lists by extending one of them
boysNames = ['John', 'Jim', 'Alex', 'Fred']
girlsNames = ['Sarah', 'Alex', 'Pat', 'Mary']
boysNames = boysNames + girlsNames
print(boysNames)
```

```python
fruits = ['Strawberry', 'Lemon', 'Orange', 'Raspberry', 'Cherry']
print(fruits)

fruits[0] = 'Apple'
fruit = 'Melon'
fruits[1] = fruit
fruits[2] = 'Raspberry'
fruits[3] = fruits[4]
fruits[4] = 'Pineapple'

print(fruits)
```

```python
# A program to demonstrate list slicing
fruits = ['apple', 'pear', 'orange', 'banana', 'kiwi']

print(fruits[1:3])  # ['pear', 'orange']
print(fruits[2:4])  # ['orange', 'banana']
print(fruits[2:5])  # ['orange', 'banana', 'kiwi']

print(fruits[1:])   # ['pear', 'orange', 'banana', 'kiwi']
print(fruits[:5])   # ['apple', 'pear', 'orange', 'banana', 'kiwi']
```
