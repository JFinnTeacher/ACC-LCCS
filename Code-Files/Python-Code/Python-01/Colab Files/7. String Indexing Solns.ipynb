# Exercise 3

```python
# Program to demonstrate the use of common string methods

pangram = "the quick brown fox jumped over the lazy dog!"

print(pangram.capitalize())
print(pangram.upper())
print(pangram.isupper())
print(pangram.islower())
print(pangram.count("o"))
print(pangram.find("fox"))
print(pangram.find("Fox"))
print(pangram.find("the"))
print(pangram.find("z"))
```
